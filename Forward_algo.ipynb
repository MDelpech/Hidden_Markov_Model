```python
import numpy as np

# Forward Algorithm for Hidden Markov Models
```

## Le forward algorithm calcule les quantités

$$
\alpha_t(i) = \mathbb{P}(Y_1,\dots,Y_t, X_t=i)
$$

où :
- \(X_t\) est l'état caché au temps \(t\),
- \(Y_t\) est l'observation au temps \(t\).

L'idée est de calculer récursivement la probabilité partielle de la séquence observée jusqu'au temps \(t\).

```python
def forward_algo(A, B, pi, Y):
    """
    Compute the forward probabilities for a Hidden Markov Model.

    Parameters
    ----------
    A : np.ndarray
        Transition matrix of shape (N, N)
    B : np.ndarray
        Emission matrix of shape (N, M)
    pi : np.ndarray
        Initial distribution of shape (N,)
    Y : list or np.ndarray
        Sequence of observations

    Returns
    -------
    alpha : np.ndarray
        Forward probability matrix of shape (N, T)
    """
    T = len(Y)
    N = A.shape[0]
    
    assert A.shape == (N, N), "A must be a square transition matrix"
    assert B.shape[0] == N, "B must have N rows"
    assert pi.shape[0] == N, "pi must have length N"

    alpha = np.zeros((N, T))

    # Initialisation
    alpha[:, 0] = pi * B[:, Y[0]]

    # Récursion
    for t in range(1, T):
        for j in range(N):
            alpha[j, t] = np.dot(alpha[:, t-1], A[:, j]) * B[j, Y[t]]

    return alpha
```

##  Likelihood function

La vraisemblnace de la séquence d'observation est obtenue en sommant les $\alpha_T(i)$ :

$$
\mathbb{P}(Y_1,\dots,Y_T) = \sum_{i=1}^N \alpha_T(i)
$$

```python
def likelihood(A,B,pi,Y) :
    alpha = forward_algo(A,B,pi,Y)
    return np.sum(alpha[:,-1])
```

## Example 

```python
A = np.array([
    [0.7, 0.3],
    [0.4, 0.6]
])

B = np.array([
    [0.9, 0.1],   
    [0.2, 0.8]   
])

pi = np.array([0.6, 0.4])
Y = [0, 1, 0]

alpha = forward_algo(A, B, pi, Y)

print("Forward matrix alpha:")
print(alpha)

print("\nLikelihood of the observation sequence:")
print(likelihood(A, B, pi, Y))
```

La somme des composantes de la dernière colonne de $\alpha$ donne :

$$
\mathbb{P}(Y_1,\dots,Y_T)
$$

c’est-à-dire la vraisemblance de la séquence observée sous le modèle HMM choisi.
